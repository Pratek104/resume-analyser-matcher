```python
from langchain_core.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Initialize the LLM with Groq API
llm = ChatGroq(
    temperature=0,
    groq_api_key=os.getenv('GROQ_API_KEY'),
    model_name="llama-3.1-70b-versatile"
)

# Import your preprocessed data
from pdftotxt import limited_text  # The extracted text from the resume PDF
from webbase import res  # The job portal text

print(res)


# Improved prompt for comparing resume and job description
prompt_extract = PromptTemplate.from_template(
    """
    I will provide two texts:
    
    1. The job listing from the company's job portal: 
    "{res}"
    
    2. My resume: 
    "{limited_text}"
    
    Context:
    - Please compare the job description and my resume.
    - Focus on the following sections:
        - Qualifications
        - Skills
        - Experience
    
    Task:
    - Provide a detailed comparison of how well my resume matches the job requirements.
    - Rate the match as a percentage based on the alignment of skills, qualifications, and experience.
    """
)


chain_extract = LLMChain(
    llm=llm,
    prompt=prompt_extract
)


res_output = chain_extract.run({
    'res': res,  
    'limited_text': limited_text 
})


print(res_output)
```
